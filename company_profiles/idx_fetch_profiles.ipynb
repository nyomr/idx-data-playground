```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import json
import pandas as pd
```

```python
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
```

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(
    ChromeDriverManager().install()), options=options)
url = "https://www.idx.co.id/primary/ListedCompany/GetCompanyProfiles?emitenType=s&start=0&length=9999"
driver.get(url)

raw = driver.find_element("tag name", "pre").text
data = json.loads(raw)
```

```python

df = pd.DataFrame(data["data"])
```

```python
dfx = df[["Alamat", "BAE", "Industri", "SubIndustri", "Email", "Fax", "KegiatanUsahaUtama", "KodeEmiten", "NamaEmiten", "NPWP", "PapanPencatatan", "Sektor", "SubSektor", "TanggalPencatatan", "Telepon", "Website"]]
```

```python
total_count = dfx.shape[0]
print("Total data:", total_count)
```

```python
filename = "profil_perusahaan.csv"
dfx.to_csv(filename, index=False, encoding="utf-8-sig")
print(f"Data has been saved to {filename}")
```
